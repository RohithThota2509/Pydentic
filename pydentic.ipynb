# Model with optional fields.

```python
from typing import Optional
class Employee(BaseModel):
    id:int
    name:str
    dept:str
    salary:Optional[float]=None
    is_active:Optional[bool]=True

emp1=Employee(id=1,name="Mahesh",dept="CSE")
print(emp1)
```

```python
emp2=Employee(id=2,name="Kalyani",dept="ECE",salary=600000)
print(emp2)
```

Using List of Strings

```python
from typing import List
from pydantic import BaseModel

class ClassRoom(BaseModel):
    room_no:str
    student:List[str]
    capacity:int

std1=ClassRoom(room_no="101",student=["mahesh","kalyani","Raji","PullaRao","Rohan","Kethana","Rohith","Sathyanarayana","Kalavathi"],capacity=9)
print(std1)
```

```python
try:
    std2=ClassRoom(room_no="101",student=["mahesh","kalyani","Raji","PullaRao","Rohan","Kethana","Rohith","Sathyanarayana","Kalavathi",123],capacity=10)
    print(std2)
except ValueError as e:
    print(e)
```

Model with nested models

create complex structures with nested models:

```python
from pydantic import BaseModel
class Adress(BaseModel):
    street:str
    city:str
    zipcode:int

class Customer(BaseModel):
    customer_id:int
    name:str
    adress:Adress

cust=Customer(
    customer_id=1,
    name="Ramesh",
    adress={"street":"lane1","city":"Hyd","zipcode":"101"}
)
print(cust)
```

pydantuc fields:customization and constrains

```python
from pydantic import BaseModel,Field
class Item(BaseModel):
    name:str=Field(min_len=2,max_len=10)
    price:float=Field(gt=0,le=1000)
    quantity:int=Field(gt=0)

item1=Item(name="Book12345678",price=29.99,quantity=10)
print(item1)
```

```python
print(Item.model_json_schema())
```

```python
from pydantic import BaseModel
class User(BaseModel):
    name:str=Field(...,description="this is users name which must be a string for the usage")
    age:int=Field(default=17)

user1=User(name="Maikanth")
print(user1)
```
